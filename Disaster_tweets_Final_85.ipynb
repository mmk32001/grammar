```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
train_data = pd.read_csv("/content/drive/MyDrive/train.csv")
test_data =pd.read_csv("/content/drive/MyDrive/test.csv")
```

```python
train_data = train_data.drop(columns = ['id', 'keyword','location'])
test_data = test_data.drop(columns = ['id', 'keyword','location'])
```

```python
train_data['text'] = train_data['text'].str.replace(r'\d+',"", regex=True)
test_data['text'] = test_data['text'].str.replace(r'\d+',"",regex=True)

train_data['text']  = train_data['text'].str.replace(r'http[s]?://\S+',"",regex=True)
test_data['text']  = test_data['text'].str.replace(r'http[s]?://\S+',"",regex=True)

train_data['text'] = train_data['text'].str.replace(r'[^\w\s]',"",regex=True)
test_data['text'] = test_data['text'].str.replace(r'[^\w\s]',"",regex=True)
```

```python
train_data
```

```python
pip install datasets
```

```python
from datasets import Dataset, DatasetDict
from sklearn.model_selection import train_test_split

train_df, eval_df = train_test_split(train_data, train_size=0.97, stratify = train_data.target)   # stratify means equal distribution of target values in train and eval data
```

```python
my_dataset = DatasetDict({
    "train":Dataset.from_pandas(train_df),
    "eval": Dataset.from_pandas(eval_df)
})
```

```python
my_dataset['train'][0]
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
def tokenize_fun(data):
  return tokenizer(data['text'],truncation=True)

tokenized_data = my_dataset.map(tokenize_fun, batched=True)
```

```python
tokenized_data = tokenized_data.remove_columns(["text","__index_level_0__","token_type_ids"])
tokenized_data = tokenized_data.rename_column("target","labels")
tokenized_data.set_format("torch")
print(tokenized_data)
```

```python
pip install evaluate
```

```python
from torch.utils.data import DataLoader
from transformers import DataCollatorWithPadding
import evaluate
import numpy as np

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

train_dataset = tokenized_data["train"].shuffle(seed=42)
eval_dataset = tokenized_data["eval"].shuffle(seed=42)

train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=16, collate_fn=data_collator)
eval_dataloader = DataLoader(eval_dataset, batch_size=16,collate_fn=data_collator)
```

```python
from torch.optim import AdamW
```

```python
num_epochs = 56
num_training_steps = num_epochs * len(train_dataloader)
```

```python
import torch

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
```

```python
torch.cuda.empty_cache()
device
```

```python
from transformers import AutoConfig
config = AutoConfig.from_pretrained("bert-base-uncased",num_labels=2)

classifier = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", config=config)
classifier = classifier.to(device)

optimizer = AdamW(classifier.parameters(), weight_decay=0.01)
scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr = 0.0000001, steps_per_epoch=len(train_dataloader), epochs=56)
```

```python

for epoch in range(num_epochs):
  j=0
  total_train_loss = 0.0
  total_eval_loss = 0.0
  print(f"Epoch : {epoch+1}")
  for batch in train_dataloader:
    
    batch = {k: v.to(device) for k,v in batch.items()}  #this will form dictionary like structure.Here k is "labels","input_ids","attention_masks"
 
    outputs = classifier(**batch)
    loss = outputs.loss
    total_train_loss+=loss
    
    loss.backward()  #compute gradient of the weights with respect to loss
    scheduler.step()
    optimizer.step()  # optimize means update the weights of the network based on computed gradients

    optimizer.zero_grad()
    j=j+1

  classifier.eval()
  with torch.no_grad():
    for batch in eval_dataloader:
      batch = {k: v.to(device) for k,v in batch.items()}
      outputs = classifier(**batch)
      loss = outputs.loss
      total_eval_loss += loss
    
  
  lr=scheduler.get_lr()
  print("lr : " + str(lr))
  run["training/epoch/lr"].log(lr)

  print(f"TRAIN LOSS: "+ str(total_train_loss/len(train_dataloader)))
  print(f"VALI LOSS: "+ str(total_eval_loss/len(eval_dataloader)))

  run["training/epoch/loss_train"].log(total_train_loss/len(train_dataloader))
  run["training/epoch/loss_eval"].log(total_eval_loss/len(eval_dataloader))

  
```

```python
import evaluate

metric = evaluate.load("accuracy")
classifier.eval()
sum=0.0
for batch in eval_dataloader:
  batch = {k: v.to(device) for k,v in batch.items()}
  with torch.no_grad():
    outputs = classifier(**batch)     # no gradient calculation
  
  loss = outputs.loss
  run[f"testing/batch/loss_test"].log(loss)
  sum += loss 
  logits = outputs.logits
  predictions = torch.argmax(logits, dim=-1)      # argmax is applied along last dimension
  metric.add_batch(predictions=predictions, references=batch["labels"])

error = (sum/len(eval_dataloader))
print("Validation Loss : " + str(error))


x = metric.compute()
print("Validation Accuracy : " + str(x))

run['valid_acc'] = x
run['valid_loss'] = error
```

```python
import evaluate

metric = evaluate.load("accuracy")
classifier.eval()
sum=0.0
for batch in train_dataloader:
  batch = {k: v.to(device) for k,v in batch.items()}
  with torch.no_grad():
    outputs = classifier(**batch)     # no gradient calculation
  
  loss = outputs.loss
  sum += loss 
  logits = outputs.logits
  predictions = torch.argmax(logits, dim=-1)      # argmax is applied along last dimension
  metric.add_batch(predictions=predictions, references=batch["labels"])

error = (sum/len(train_dataloader))
print("Train Error : " + str(error))

x = metric.compute()
print("Train Accuracy : " + str(x))


run['Train_acc'] = x
run['Train_loss'] = error
```
